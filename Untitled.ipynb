```python
import pandas as pd
import seaborn as sns
import warnings
warnings.simplefilter('ignore')
Code = pd.read_csv('Code/TOPIXCore.csv', names=['0', 'Code'])['Code'].to_list()
Code += pd.read_csv('Code/TOPIXLarge.csv', names=['0', 'Code'])['Code'].to_list()
dataX = pd.DataFrame()
for code in Code:
    try:
        pd.read_csv(f'Price/TOPIX/'+str(code).replace('.T', '') + '.csv', index_col='Date')
    except:
        print(code)
    else:
        dataX[str(code)] = pd.read_csv(f'Price/TOPIX/'+str(code).replace('.T', '') + '.csv', index_col='Date')['rsi9']
```

```python
import matplotlib.pyplot as plt



# 相関行列の計算
corr_matrix = dataX.corr()

# ヒートマップの描画
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=False, fmt=".2f", cmap="coolwarm", square=True, cbar=True)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig('RSI_corr.png')
plt.show()
```
