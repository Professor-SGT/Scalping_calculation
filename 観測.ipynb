```python
import pandas as pd
tickers = pd.read_csv('Code/JPNK400.csv', index_col=0)['0'].values
tickers
```

```python
data = pd.DataFrame()
for ticker in tickers:
    
    try:
        daily_df = pd.read_csv('Price/TOPIX/{}.csv'.format(ticker.replace('.T', '')), index_col='Date', parse_dates=True)
    except Exception as e:
        print(e)
        continue
    daily_df['date'] = pd.to_datetime(daily_df.index).normalize()
    daily_df = daily_df.set_index('date')
    data[ticker] = daily_df['rsi9']
    #data = pd.concat([data, daily_df[['rsi9']]], axis=1)
data
```

```python
data
```

```python
for i, row in data.iterrows():
    print((row > 80).sum(axis=0))
```

```python
data['ratio'] = (data > 80).sum(axis=1) / data.shape[1]
```
